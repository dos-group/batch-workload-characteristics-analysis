```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Basic cosine function
y_basic = np.cos(x)

# Plotting the basic cosine function
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y_basic, label='cos(x)')
plt.title('Basic Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.legend()

# Demonstrating periodicity
plt.subplot(2, 2, 2)
plt.plot(x, y_basic, label='cos(x)')
plt.plot(x, np.cos(x + 2*np.pi), 'r--', label='cos(x + 2π)')
plt.title('Periodicity of Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.legend()

# Phase shift
plt.subplot(2, 2, 3)
plt.plot(x, y_basic, label='cos(x)')
plt.plot(x, np.cos(x + np.pi/4), 'g--', label='cos(x + π/4)')
plt.title('Horizontal Shift (Phase Shift)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.legend()

# Amplitude scaling
plt.subplot(2, 2, 4)
plt.plot(x, y_basic, label='cos(x)')
plt.plot(x, 2*np.cos(x), 'm--', label='2*cos(x)')
plt.title('Vertical Scaling (Amplitude Change)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

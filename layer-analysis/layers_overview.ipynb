```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Define the layers with additional info and adjust the order
left_layers_with_info = [
    ('Data Layer', 'Graybox Layer'),
    ('User Application Layer', 'Blackbox Layer'),
    ('Big Data Framework Layer', 'Whitebox Layer'),
    ('Virtualisation Layer', 'Graybox Layer'),
    ('Hardware Layer', 'Blackbox Layer')
]
right_layer_info = 'Performance Layer'
right_layer_type = 'Graybox Layer'

# Define colors for each type of layer with Blackbox layers darker than Graybox layers
color_map = {
    'Blackbox Layer': 'dimgray',
    'Graybox Layer': 'darkgray',
    'Whitebox Layer': 'white'
}

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

# Draw the left layers with additional info
for i, (layer, info) in enumerate(left_layers_with_info):
    rect = patches.FancyBboxPatch((1, 8 - 1.5*i), 3, 1, boxstyle="round,pad=0.1", edgecolor='black', facecolor=color_map[info])
    ax.add_patch(rect)
    ax.text(1.5, 8.75 - 1.5*i, layer, verticalalignment='center', fontsize=10)
    ax.text(1.5, 8.25 - 1.5*i, f"({info})", verticalalignment='center', fontsize=8, color='black')

# Draw the right layer with additional info
rect = patches.FancyBboxPatch((6, 4), 3, 1, boxstyle="round,pad=0.1", edgecolor='black', facecolor=color_map[right_layer_type])
ax.add_patch(rect)
ax.text(6.5, 4.5, right_layer_info, verticalalignment='center', fontsize=10)
ax.text(6.5, 4.0, f"({right_layer_type})", verticalalignment='center', fontsize=8, color='black')

# Draw the arrows from left layers to right layer
for i in range(len(left_layers_with_info)):
    ax.annotate('', xy=(6, 4.5), xytext=(4, 8.5 - 1.5*i), 
                arrowprops=dict(arrowstyle="->", lw=1.5))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Layers and their parameters
layers = ['VM / Container / Cloud', 'Hardware', 'JVM', 'Big Data Framework', 'Data', 'User Application', '(Historical) Performance']
blackbox_counts = np.array([5, 10, 0, 0, 1, 0, 3])
whitebox_counts = np.array([1, 0, 3, 13, 2, 4, 15])
greybox_counts = np.array([0, 0, 0, 0, 0, 0, 0])  # Update if greybox parameters are specified

# Plotting setup
ind = np.arange(len(layers))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 8))
rects1 = ax.bar(ind - width/2, blackbox_counts, width, label='Blackbox', color='black')
rects2 = ax.bar(ind + width/2, whitebox_counts, width, label='Whitebox', color='grey')

# Add text for labels, title, and custom x-axis tick labels
ax.set_ylabel('Number of Parameters')
ax.set_title('Destribution of Parameters by Layer and Type')
ax.set_xticks(ind)
ax.set_xticklabels(layers, rotation=45, ha="right")
ax.legend()

# Adding a grid
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)

# Function to attach a text label above each bar, displaying its height
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
